Vytvorenie grafu

```python
graph = {}
nodes = set()
edges = []
with open('linky.txt','r',) as file:
    for line in file:
        stanice = line.split(':  ')[1].strip('\n').split(';')
        linka =  line.split(':  ')[0]
        print(linka)
        for i,stanica in enumerate(stanice[:-1]):
            
            nodes.add(stanica)
            edges.append((stanice[i-1],stanica))
            if i != 0:
                try:
                    graph[stanice[i-1]][stanica] += 1
                except KeyError:
                    try:
                        graph[stanice[i-1]][stanica] = 1
                    except KeyError:
                        graph[stanice[i-1]] = {}
                        graph[stanice[i-1]][stanica] = 1
```

Základné info

```python
n = len(nodes)
m = len(edges)
print(n,m)
```
